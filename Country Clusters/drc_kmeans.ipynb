# Donor Recipient Countries - KMeans
Notebook ini menjelaskan tentang mekanisme pengkalasteran negara-negara penerima bantuan donor berdasarkan kondisi socio-demografi negara tersebut.
Sumber data berasal dari [kaggle](https://www.kaggle.com/code/ashydv/country-clustering-hierarchical-clustering-pca/notebook).

Data memiliki 10 atribut sebagai berikut:
* country       167 non-null object
* child_mort    167 non-null float64
* exports       167 non-null float64
* health        167 non-null float64
* imports       167 non-null float64
* income        167 non-null int64
* inflation     167 non-null float64
* life_expec    167 non-null float64
* total_fer     167 non-null float64
* gdpp          167 non-null int64

Kasus ini termasuk dalam Unsupervised Learning karena kita tidak memiliki label kelayakan penerima donor pada data negara di dataset. Kita akan menambahkan satu kolom baru pada dataset yang berisikan nilai apakah suatu negara adalah pemberi bantuan atau penerima bantuan.

```python
import pandas as pd

# Ambil dataset yang akan menjadi target perlatihan model:
dataset_awal = pd.read_csv('/content/Country-data.csv')
dataset_awal.head()
```

Periksa dataset dengan .info()

```python
dataset_awal.info()
```

Dataset terdiri dari 10 kolom. Perhatikan ada satu kolom yang tidak bernilai numerik, yaitu kolom nama negara (country). Kolom ini tidak kita perlukan dalam pelatihan model sehingga bisa kita hapus.

```python
# Hapus kolom negara (country)
dataset = dataset_awal.drop('country', axis=1)
dataset.head()
```

Periksa batasan (range) nilai masing-masing kolom dengan fungsi .describe()

```python
dataset.describe().T
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

Perhatikan terdapat perbedaan besar batasan nilai maksimum dan minimum di antara kolom-kolom pada data, ini akan menimbulkan bias pada proses pembelajaran model. Kita akan mengubah (menskalakan) nilai-nilai pada masing-masing kolom dengan metode standarisasi menggunakan fungsi fit_transform.

```python
# Standarisasikan nilai kolom
scaler = StandardScaler()
dataset_std = scaler.fit_transform(dataset)
```

Periksa hasil standarisasi nilai kolom-kolom.

```python
pd.DataFrame(data=dataset_std, columns=dataset.columns).describe().T
```

Hasilnya semua kolom memiliki nilai rata-rata (mean) dan deviasi standar (std) mendekati 0 dan 1. Data siap digunakan untuk pelatihan model.

```python
# Import library KMeans
from sklearn.cluster import KMeans
```

Sebelum pelatihan, kita akan menentukan berapa jumlah kluster (hyperparameter) yang akan dibuat. Pada kasus ini hanya 2 kluster.

```python
# Siapkan algoritme pembelajaran K-Means:
model = KMeans(2)
```

Lakukan pelatihan model dengan fungsi .fit_predict()

```python
# Lakukan pelatihan cluster dengan K-Means untuk 2 cluster
clusters = model.fit_predict(dataset_std)
```

Lihat kluster yang terbentuk.

```python
print(clusters)
```

Hasilnya berupa deretan nomor-nomor cluster pada masing-masing negara pada dataset_std. Terdapat dua kluster dengan kode 0 dan 1 untuk masing-masing negara. Data ini akan kita tambahkan sebagai kolom baru di dataset.

```python
# Tambahkan kolom baru bernama “cluster” pada dataset awal
dataset_awal['cluster'] = clusters
```

Lihat hasil penambahan kolom baru.

```python
dataset_awal[['country', 'cluster']].head(5)
```

Kita akan coba tampilkan persebaran kluster dengan scatterplot.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["font.size"] = 7
plt.figure(figsize=(5,2), dpi=200)

p = sns.scatterplot(data=dataset_awal,
                    x='child_mort',
                    y='life_expec',
                    hue='cluster',
                    size='income',
                    sizes=(10,200),
                    style='cluster',
                    alpha=0.7,
                    markers=['s', 'o'])

p.set_xlabel('Child Mortality')
p.set_ylabel('Life Expectancy')

h,l = p.get_legend_handles_labels()
plt.legend(h[1:3],l[1:3], loc=2, bbox_to_anchor=(0.8, 0.95), title='cluster')
```
